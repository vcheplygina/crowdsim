```python
import pandas as pd
import numpy as np
```

```python
df_raw = pd.read_csv("ISIC-2017_Training_Part3_GroundTruth.csv")
df_raw
```

```python
df_labels = df_raw[["image_id","melanoma"]]
```

```python
df_labels["cancer"] = df_labels["melanoma"]
```

```python
df_labels = df_labels.drop(columns="melanoma")
df_labels
```

```python
## Randomize the order (to deal with clinic bias and maybe more)
df_random = df_labels.sample(frac=1)
```

```python
df_random
```

```python
## Save this to csv (maybe not needed)
df_random.to_csv("ISIC-2017_Training_Part3_GroundTruth_Random.csv")
```

```python
## Pick 100 random images where cancer = 1, and 100 where cancer = 0
## First part with .loc is a "where-statement", and sample(100) takes 100 items at random from the random list
df_sample_cancer = df_random.loc[(df_random["cancer"] == 1)].sample(100)
df_sample_cancer

## Same as above, but with no_cancer
df_sample_no_cancer = df_random.loc[(df_random["cancer"] == 0)].sample(100)
df_sample_no_cancer
```

```python
## Concat the two sample groups:
df_sample = pd.concat([df_sample_cancer,df_sample_no_cancer])
df_sample
```

```python
## Check that there are exactly 200 image_ids (no duplicates) and only 2 classes (cancer / no-cancer)
df_sample.nunique()
```

```python
## Save the sample data in a csv file
df_sample.to_csv("ISIC-2017_anchor_images.csv")
```
